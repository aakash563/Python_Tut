* Compute primes from 1 to 100

```python
def factors(n):
    factorlist = []
    for i in range(1,n+1):
        if n%i == 0:
            factorlist.append(i)
    return factorlist
```

```python
def prime(n):
    return(factors(n)==[1,n])
```

```python
def prime1(n):
    return(len(factors(n))==2)
```

```python
primelist = []
for i in range(1,101):
    if(prime(i)):
        primelist.append(i)
print(primelist)
```
